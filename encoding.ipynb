```python
from sklearn.preprocessing import OrdinalEncoder
import numpy as np
import pandas as pd
```

# upsampling/OverSampling

```python
df
```

```python
X=df.drop(['Occurrence (Y/N)'],axis=1)
```

```python
y=df['Occurrence (Y/N)']
```

```python
y
```

```python
X
```

```python
df['Occurrence (Y/N)'].value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
over=  RandomOverSampler(sampling_strategy=0.30)
```

```python
X_new,y_new=over.fit_resample(X,y)
```

```python
X_new
```

```python
y_new.value_counts()
```
